**Data Preprocessing**



*   Ubah variable `created_at` menjadi index



```python
house = pd.read_csv(dataC, parse_dates = True, index_col = "created_at", low_memory = False)
```



*   Missing Value



```python
house.isnull().sum()
```

Berhubung di website tidak setiap data rumah tercantum data yang lengkap, untuk variabel `garasi` dan
`carport` jika tidak tercantum angka apapun maka akan diganti dengan nilai 0.

```python
house["garasi"].fillna(0, inplace = True)
house["carport"].fillna(0, inplace = True)
```

```python
house.isnull().sum()
```

Sisanya yaitu variabel listrik dan hadap tidak dilakukan apa-apa terhadap missing value yang ada, karena
ada kemungkinan penjual rumah sengaja tidak mengisi atau lupa menginput data. Tidak bisa diganti dengan
nilai 0 karena tidak mungkin sebuah rumah tidak memiliki listrik dan tidak ada hadapnya kemana.



*   Duplicated Data



```python
house.duplicated().sum()
```



*   ubah tipe data



```python
house['garasi'] = house['garasi'].astype('int64')
house['carport'] = house['carport'].astype('int64')
house['lokasi'] = house['lokasi'].astype('category')
house['sertifikat'] = house['sertifikat'].astype('category')
house['listrik'] = house['listrik'].astype('category')
house['hadap'] = house['hadap'].astype('category')
house.dtypes
```



*   Menghapus Variable tidak penting



```python
house.drop("Unnamed: 14", axis=1, inplace=True)
house.drop("Unnamed: 15", axis=1, inplace=True)
house.drop("Unnamed: 16", axis=1, inplace=True)
house.drop("Unnamed: 17", axis=1, inplace=True)
house.drop("Unnamed: 18", axis=1, inplace=True)
house.drop("Unnamed: 19", axis=1, inplace=True)
house.drop("Unnamed: 20", axis=1, inplace=True)
house.drop("Unnamed: 21", axis=1, inplace=True)
house.drop("Unnamed: 22", axis=1, inplace=True)
house.drop("Unnamed: 23", axis=1, inplace=True)
house.drop("Unnamed: 24", axis=1, inplace=True)
house.drop("Unnamed: 25", axis=1, inplace=True)
```
